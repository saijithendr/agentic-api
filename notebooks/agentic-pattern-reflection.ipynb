# Reflection Agentic Pattern



* LLM evaluates its own outputs to suggest modifications, improvements and refinements to enhance the final result.

> Generate block
- generate response to user prompt

> Reflect block
- output from generate block is used to evaluate or review the response and get the feedback

The reflection loop can be divided into the following four steps:

 The LLM generates an output (let's call it the  )

 The LLM corrects the ouptut generated in the previous step (let's call it the  )

 The LLM takes the corrections and uses them to modify the original output accordingly.

 A new iteration starts.
